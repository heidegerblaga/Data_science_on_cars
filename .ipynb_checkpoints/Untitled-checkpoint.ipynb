```python
import pandas as pd

file = 'cars.pickle'
cars = pd.read_pickle(file)

print(cars.describe())
```

```python
import matplotlib.pyplot as plt


plt.hist(cars['cena'], bins=1000, edgecolor='k')
plt.title('Histogram Rozkładu Kolumny')
plt.xlabel('Cena')
plt.ylabel('Liczność')
plt.gca().get_xaxis().set_major_formatter(plt.ScalarFormatter(useMathText=False))
plt.gca().get_xaxis().get_major_formatter().set_scientific(False)
plt.xlim(0, 300000)
plt.show()
```

```python
plt.hist(cars['id_marki'], bins=100, edgecolor='k')
plt.title('Histogram Rozkładu Kolumny')
plt.xlabel('id_marki')
plt.ylabel('Liczność')
```

```python
import matplotlib.pyplot as plt


plt.pie(cars.groupby(['id_marki']).count()['id'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')

plt.figure(figsize=(8, 8))  # Szerokość i wysokość figury w calach

plt.title('Wykres Kołowy')

plt.show()
```
