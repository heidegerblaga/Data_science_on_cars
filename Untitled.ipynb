```python
import pandas as pd

file = 'cars.pickle'
cars = pd.read_pickle(file)

cars[cars['marka']==65]
```

```python
import matplotlib.pyplot as plt


plt.hist(cars['cena'], bins=1000, edgecolor='k')
plt.title('Histogram RozkÅ‚adu Kolumny')
plt.xlabel('Cena')
plt.ylabel('LicznoÅ›Ä‡')
plt.gca().get_xaxis().set_major_formatter(plt.ScalarFormatter(useMathText=False))
plt.gca().get_xaxis().get_major_formatter().set_scientific(False)
plt.xlim(0, 300000)
plt.show()
```

```python
plt.hist(cars['id_marki'], bins=100, edgecolor='k')
plt.title('Histogram RozkÅ‚adu Kolumny')
plt.xlabel('id_marki')
plt.ylabel('LicznoÅ›Ä‡')
```
